```python
##Importing the dependencies
import cv2
import dlib
import  numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
##Funtion to draw bounding box over detected faces
def show_detection(image,faces):
    for face in faces:
        cv2.rectangle(image,(face.rect.left(),face.rect.top()),(face.rect.right(),face.rect.bottom()),(255,0,0),4)
    return image
```

```python
##Function to show detected faces
def show_with_matplotlib(image,title):
    image = image[:,:,::-1]
    plt.imshow(image)
    plt.title(title)
    plt.axis('off')
```

```python
##initializing the dlib's cnn based detector
detector = dlib.cnn_face_detection_model_v1('models/mmod_human_face_detector.dat')

##Loading the image
img = cv2.imread('faces.jpg')

##Detecting faces in an image
faces = detector(img,0)

##Drawing rectangles over detected faces
image = show_detection(img.copy(),faces)

##showing the detected faces in an image
fig = plt.figure(figsize=(10,8))
plt.suptitle('Face detection with dlib CNN',fontsize=14,fontweight='bold')
fig.patch.set_facecolor('silver')
show_with_matplotlib(image,'Dlib CNN: '+str(len(faces)))
plt.show()
```
