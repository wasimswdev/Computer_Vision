```python
import face_recognition
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def show_detections(image,rects):
    for face in rects:
        top,right,bottom,left = face
        cv2.rectangle(image,(left,top),(right,bottom),(255,0,0),8)
    return image
```

```python
def show_with_matplotlib(image,title,pos):
    ax = plt.subplot(1,2,pos)
    plt.imshow(image)
    plt.title(title)
    plt.axis('off')
```

## Face detection using face recognition library using hog features

```python
image = cv2.imread('faces.jpg')
rgb = image[:,:,::-1]

rects1 = face_recognition.face_locations(rgb,0,'hog')
rects2 = face_recognition.face_locations(rgb,1,'hog')


faces1 = show_detections(rgb.copy(),rects1)
faces2 = show_detections(rgb.copy(),rects2)
```

## Face detection using face recognition library using CNN features

```python
##getting bounding boxes from fr library
rects1 = face_recognition.face_locations(rgb,0,'cnn')
rects2 = face_recognition.face_locations(rgb,1,'cnn')

##drawing rectangles over face using bbox
faces1 = show_detections(rgb.copy(),rects1)
faces2 = show_detections(rgb.copy(),rects2)
```

## show detections

```python
##showing with matplotlib
fig = plt.figure(figsize=(12,8))
plt.suptitle('Face detection using face recongition library(CNN)',fontsize=14,fontweight='bold')
fig.patch.set_facecolor('silver')
show_with_matplotlib(faces1,'not Upsampled :'+str(len(rects1)),1)
show_with_matplotlib(faces2,'Upsampled: '+str(len(rects2)),2)
plt.show()
```
